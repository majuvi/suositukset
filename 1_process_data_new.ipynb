```python
# Get stocks last recommendation, replace by - if not updated in a quarter (end follow-up)
fix_last = df.groupby('Osake', as_index=False)[['Pvm', 'Suositus']].last()
fix_last.loc[fix_last['Pvm'] <= pd.Timestamp('2021-6-30'), 'Suositus'] = '-'
df = df.merge(fix_last, how='left', on=['Osake', 'Pvm'], suffixes=('', '_korjaa'))
df.loc[~df['Suositus_korjaa'].isnull(), 'Suositus'] = df['Suositus_korjaa']
```

```python
df = df[df['Pvm'] <= pd.Timestamp('2021-12-30')]
```
